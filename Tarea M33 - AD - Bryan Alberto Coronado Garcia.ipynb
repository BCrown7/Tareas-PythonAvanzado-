```python
# Eliminiación de datos nulos para mejor valoración.

df = df.dropna()
df.isnull().sum()
```

```python
# Cambié la columna de año a Integer debido a que resultaba más útil su uso de esta forma que en 'Datetime'

df['Year'] = df['Year'].astype('int32')

df.info()
```

#### Ventas

```python
# Ventas por año (Millones De Dolares)

df.groupby('Year')['Global_Sales'].sum().plot.bar(grid=True, alpha=0.5)
```

```python
# El videojuego más vendido de cada año (Millones De Dolares)

df.groupby(['Year', 'Name'])['Global_Sales'].sum().groupby(level=0).nlargest(1).reset_index(level=0, drop=True).plot.bar(grid=True, alpha=0.5)
```

```python
# Plataformas con más ingresos por año (Millones De Dolares)

df.groupby(['Year', 'Platform'])['Global_Sales'].sum().groupby(level=0).nlargest(1).reset_index(level=0, drop=True).plot.bar(grid=True, alpha=0.5)
```

#### Gráfico de comparación de ventas entre géneros por año

```python
ventas_genero = df.pivot_table(
    index='Year',
    columns='Genre',
    values='Global_Sales'
)

plt.figure(figsize=(12, 7))
ventas_genero.plot(kind='bar', stacked=True, ax=plt.gca())

plt.title('Ventas Globales por Género Cada Año')
plt.xlabel('Año')
plt.ylabel('Ventas Globales (Millones De Dolares)')
plt.xticks(rotation=50)
plt.legend(title='Géneros',  bbox_to_anchor=(1,1), loc='upper right')
plt.grid(axis='y', linestyle='-', alpha=0.5)
plt.show()

#plt.savefig('Ventas Globales Por Genero Cada Año.png', dpi=400)
```

#### Insights: 2 Gráficos adicionales interesantes

```python
# Top 10 distribuidores con mayor cantidad en ventas de videojuegos

df.groupby(['Publisher'])['Global_Sales'].sum().nlargest(10).sort_values(ascending=True).plot.barh(
    ylabel='Distribuidor',
    xlabel='Millones De Dolares',
    title='Distribuidores con mas ventas en la historia',
    rot=0,
    grid=True,
    alpha=0.8,
    figsize= (11,6),
    color='green')

plt.savefig('Distribuidores con mas ventas en la historia',
            dpi=300,
            bbox_inches ="tight",
            pad_inches = 0.3,
            orientation ='portrait')
```

```python
# Top 10 videojuegos más vendidos de la historia (en millones de ventas)

df.groupby(['Name'])['Global_Sales'].sum().nlargest(10).sort_values(ascending=True).plot.barh(
    ylabel='Videojuegos',
    xlabel='Millones De Dolares',
    title='Los 10 videojuegos más vendidos de la historia',
    rot=0,
    grid=True,
    alpha=0.8,
    figsize= (11,6),
    color='red')

plt.savefig('Los videojuegos más vendidos de la historia',
            dpi=300,
            bbox_inches ="tight",
            pad_inches = 0.3,
            orientation ='portrait')
```

#### Crecimiento

```python
""" Los últimos cuatro años disponibles en la metadata constan del 2013 al 2016, puesto que
    después del 2017 los datos son carentes, por lo que estos serán considerados para mostrar el
    crecimiento como los años más recientes """

df_eda = df.loc[df['Year'] > 2012]
df_eda1 = df_eda.loc[df['Year'] < 2017]

df_eda1.sample(10)
```

```python
# 1) Qué género ha tenido el mayor y menor crecimiento:

pd.crosstab(df_eda1['Year'], df_eda1['Genre'],  normalize='index', margins=True).plot(
    ylabel='Cantidad',
    xlabel='Años',
    title='Crecimiento Por Generos',
    rot=0,
    grid=True,
    alpha=0.8,
    figsize= (12,6))

plt.savefig('Crecimiento por generos en cantidad de ventas',
            dpi=300,
            bbox_inches ="tight",
            pad_inches = 0.3,
            orientation ='portrait')
```

```python
# 2) Qué plataforma ha tenido el mayor y menor crecimiento

pd.crosstab(df_eda1['Year'], df_eda1['Platform'],  normalize='index',margins=True).plot(
    ylabel='Cantidad',
    xlabel='Años',
    title='Crecimiento Por Plataformas',
    rot=0,
    grid=True,
    alpha=0.8,
    figsize= (12,6))

plt.savefig('Crecimiento por plataformas en cantidad de ventas',
            dpi=300,
            bbox_inches ="tight",
            pad_inches = 0.3,
            orientation ='portrait')
```
